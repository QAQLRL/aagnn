```python
import matplotlib.pyplot as plt

x = [100, 200, 300, 400]
y1 = [0.260, 0.258, 0.255, 0.250]
y2 = [0.663, 0.581, 0.532, 0.496]
y3 = [0.81, 73, 0.532, 0.496]
# plt.title('')  # 折线图标题
plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示汉字
plt.xlabel('data size')  # x轴标题
plt.ylabel('MAE ev/atom')  # y轴标题
plt.plot(x, y1, marker='o', markersize=3)  # 绘制折线图，添加数据点，设置点的大小
plt.plot(x, y2, marker='o', markersize=3)
plt.plot(x, y3, marker='o', markersize=3)
for a, b in zip(x, y1):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)  # 设置数据标签位置及大小
for a, b in zip(x, y2):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=10)

plt.legend(['pretained', 'without-pretrained','random forest'])  # 设置折线名称

plt.show()  # 显示折线图

```
